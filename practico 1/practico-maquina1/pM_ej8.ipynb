8. Implementar un programa que valide un CUIT/CUIL ingresado por teclado.

La función validar_cuit comprueba si un CUIT/CUIL es válido.
Elimina guiones, verifica que tenga 11 dígitos numéricos, calcula el dígito verificador con el algoritmo oficial (módulo 11 y pesos fijos) y lo compara con el último dígito.
Devuelve True si el CUIT/CUIL es correcto o False en caso contrario.

```python
def validar_cuit(cuit: str) -> bool:
    # Eliminar posibles guiones
    cuit = cuit.replace("-", "").strip()

    # Validar longitud y que sea numérico
    if len(cuit) != 11 or not cuit.isdigit():
        return False

    # Pesos fijos
    pesos = [5, 4, 3, 2, 7, 6, 5, 4, 3, 2]

    # Calcular suma ponderada
    suma = sum(int(cuit[i]) * pesos[i] for i in range(10))

    resto = suma % 11
    if resto == 0:
        digito_verificador = 0
    elif resto == 1:
        digito_verificador = 9
    else:
        digito_verificador = 11 - resto

    # Comparar con el último dígito del CUIT
    return digito_verificador == int(cuit[-1])
```

Programa Principal. Ingreso de cuil o cuit.

```python
if __name__ == "__main__":
    cuit = input("Ingrese un CUIT/CUIL (formato XX-XXXXXXXX-X o solo números): ")
    if validar_cuit(cuit):
        print("✅ El CUIT/CUIL es válido.")
    else:
        print("❌ El CUIT/CUIL es inválido.")
```
