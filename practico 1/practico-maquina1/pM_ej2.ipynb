2. Escribir un programa en lenguaje a elección, que permita:

a. Ingresar el nombre de un archivo con extensión .bmp

b. Valide que el archivo sea con el formato adecuado.

c. Muestre los datos de la cabecera del archivo .bmp

Se importa libreria

```python
import struct  # Para leer datos binarios
```

Función encargada de leer el archivo BMP

```python
def leer_bmp(file_path):
    try:
        with open(file_path, 'rb') as f:
            # ===== CABECERA BMP (14 bytes) =====
            bmp_header = f.read(14)
            # Desempaquetar: 2s = 2 caracteres, I = entero sin signo de 4 bytes, 4s = 4 bytes reservados
            signature, file_size, reserved, data_offset = struct.unpack('<2sI4sI', bmp_header)
            signature = signature.decode()

            # Validación de archivo BMP
            if signature != 'BM':
                print("El archivo no es un BMP válido.")
                return

            print("===== CABECERA BMP =====")
            print("Signature:", signature)        # Debe ser 'BM'
            print("FileSize:", file_size)
            print("Reserved:", reserved)
            print("DataOffset:", data_offset)    # Donde comienzan los datos de imagen

            # ===== PROPÓSITOS DE LA IMAGEN (DIB HEADER / BITMAPINFOHEADER, 40 bytes) =====
            dib_header = f.read(40)
            # Desempaquetar todos los campos según la especificación BITMAPINFOHEADER
            (size, width, height, planes, bit_count, compression, image_size,
             x_pixels_per_m, y_pixels_per_m, colors_used, colors_important) = struct.unpack('<IIIHHIIIIII', dib_header)

            print("\n===== PROPIEDADES DE LA IMAGEN =====")
            print("Size:", size)                    # Tamaño de esta sección (40 bytes)
            print("Width:", width, "pixels")
            print("Height:", height, "pixels")
            print("Planes:", planes)                # Siempre 1
            print("BitCount:", bit_count)           # Bits por píxel
            print("Compression:", compression)      # Tipo de compresión
            print("ImageSize:", image_size)         # Tamaño de la imagen comprimida
            print("XPixelsPerM:", x_pixels_per_m)  # Resolución horizontal
            print("YPixelsPerM:", y_pixels_per_m)  # Resolución vertical
            print("ColorsUsed:", colors_used)      # Colores usados si hay paleta
            print("ColorsImportant:", colors_important)  # Colores importantes

    except FileNotFoundError:
        print("Archivo no encontrado.")
    except Exception as e:
        print("Error al leer el archivo BMP:", e)

# ===== USO DEL PROGRAMA =====
nombre_archivo = input("Ingrese el nombre del archivo BMP (con extensión): ")
leer_bmp(nombre_archivo)
```
