1. Escribir un programa en lenguaje a elección, que permita:

a. Ingresar el nombre de un archivo con extensión .wav

b. Valide que el archivo sea con el formato adecuado.

c. Muestre los datos de la cabecera del archivo .wav


Librería para trabajar con datos binarios y empaquetar/desempaquetar

```python
import struct
```

Funcion que lee el archivo WAVE

```python
def leer_wav(file_path):
    try:
        # Abrir el archivo en modo binario
        with open(file_path, 'rb') as f:
            # ===== RIFF HEADER =====
            # Leer los primeros 12 bytes del archivo
            riff = f.read(12)
            # Desempaquetar los bytes: 4s = 4 caracteres, I = entero sin signo de 4 bytes
            chunk_id, chunk_size, format = struct.unpack('<4sI4s', riff)
            print("ChunkID:", chunk_id.decode())  # Debe ser 'RIFF'
            print("ChunkSize:", chunk_size)      # Tamaño total del archivo menos 8 bytes
            print("Format:", format.decode())    # Debe ser 'WAVE'

            # ===== Subchunk1 (fmt) =====
            # Leer la cabecera del chunk de formato (8 bytes)
            subchunk1_header = f.read(8)
            subchunk1_id, subchunk1_size = struct.unpack('<4sI', subchunk1_header)
            print("Subchunk1 ID:", subchunk1_id.decode())  # Debe ser 'fmt '
            print("Subchunk1 Size:", subchunk1_size)      # Tamaño del subchunk de formato

            # Leer el contenido del chunk de formato según el tamaño indicado
            fmt_chunk = f.read(subchunk1_size)
            # Desempaquetar los primeros 16 bytes que contienen los datos más importantes
            audio_format, num_channels, sample_rate, byte_rate, block_align, bits_per_sample = struct.unpack('<HHIIHH', fmt_chunk[:16])
            print("Audio Format:", audio_format)          # 1 = PCM (sin compresión)
            print("Num Channels:", num_channels)          # 1 = mono, 2 = stereo
            print("Sample Rate:", sample_rate)            # Frecuencia de muestreo (Hz)
            print("ByteRate:", byte_rate)                 # Bytes por segundo
            print("BlockAlign:", block_align)             # Bytes por muestra incluyendo todos los canales
            print("Bits Per Sample:", bits_per_sample)    # Bits por muestra

            # ===== Subchunk2 (data) =====
            # Leer la cabecera del chunk de datos (8 bytes)
            subchunk2_header = f.read(8)
            subchunk2_id, subchunk2_size = struct.unpack('<4sI', subchunk2_header)
            print("Subchunk2 ID:", subchunk2_id.decode())  # Debe ser 'data'
            print("Subchunk2 Size:", subchunk2_size)      # Tamaño del bloque de datos de audio
            print("Data: (contenido omitido, tamaño de bytes mostrado)")

    except FileNotFoundError:
        print("Archivo no encontrado")
    except Exception as e:
        print("Error al leer el archivo:", e)
```

Programa principal:
Se Solicita al usuario el nombre del archivo WAV y se lee el archivo

```python
nombre_archivo = input("Ingrese el nombre del archivo WAV (con extensión): ")
leer_wav(nombre_archivo)
```
